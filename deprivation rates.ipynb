```python
data = pd.read_csv("Household income.csv")

indices of deprivation = pd.read_csv()
```

```python
data = pd.DataFrame(data)
data
```

```python
#df = df.drop(labels=['label1', 'label2', 'label3'], axis=1)
```

```python
data = data.drop(labels=['MSOA code', 'MSOA name', 'Local authority code', 'Region code', 'Percentage of Households Below 60% of the Median Income; (after housing costs); 95% Confidence Interval Lower Limit', 'Percentage of Households Below 60% of the Median Income; (after housing costs); 95% Confidence Interval Upper Limit'], axis=1)

data
```

```python
#DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True

data_groupby = data.groupby(['Region name']).mean(numeric_only=True)

data_groupby

```

```python
data_groupby = data_groupby.sort_values('Percentage of Households Below 60% of the Median Income; (after housing costs)', ascending=False)

data_groupby
```

```python
top_10_counties = data_groupby[0:10]
```

```python
top_10_counties.plot(kind='bar')
```
