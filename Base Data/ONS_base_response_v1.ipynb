```python
# Dependencies (API key not required as ONS is an open and unrestricted API)
import json
import requests
```

```python
# Set base URL
url = "https://api.beta.ons.gov.uk/v1/datasets"
```

```python
# Get data
ons_response = requests.get(url)
ons_json = ons_response.json()

# Print response
ons_response = requests.get(url).json()
print(json.dumps(ons_response, indent=4, sort_keys=True))
```
